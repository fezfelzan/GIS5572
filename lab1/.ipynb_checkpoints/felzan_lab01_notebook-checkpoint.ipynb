## Extracting data from the Minnesota Geospatial Commons using the CKAN API

```python
gis_mngov_home = "https://resources.gisdata.mn.gov/pub/gdrs/data/"
```

```python
def PrintNavItems(mngov_page):
    
    rqst_objct = requests.get(mngov_page)
    req_text = rqst_objct.text
    split_by_newlines = req_text.split("\n")
    
    relevent_items = []
    
    for item in split_by_newlines:
        if "?C=N;O=D" in item:
            pass
        elif "/pub/gdrs/" in item:
            pass
        elif "a href=" in item:
            relevent_items.append(item)
        
    for item in relevent_items:
        firstsplit = item.split('<a href="')[1]
        sec_split = firstsplit.split('">')[0]
        print(sec_split)
```

```python
PrintNavItems(gis_mngov_home)
```

```python
user_selection = "pub/"
```

```python
new_concatenation = gis_mngov_home + user_selection
PrintNavItems(new_concatenation)
```

```python
split_stations = stationslice.split("&")
split_stations[0:5]
```

```python
station_info_url_base = "https://ndawn.ndsu.nodak.edu/station-info.html"

demo = station_info_url_base + "?" + split_stations[0]
print(demo)
```

```python
print((requests.get(demo)).text)
```

```python
def StationNameGetter(splitstation):
    full_url = station_info_url_base+"?"+splitstation
    html_text = (requests.get(full_url)).text
    frontslice = html_text.split("<h1>NDAWN Station: ")[1]
    clean_name = frontslice.split("</h1>")[0]
    return clean_name

StationNameGetter('station=78')
```

```python
station_names = {}

for station in split_stations:
    iter_station = station.replace("=", " ")
    iter_name = StationNameGetter(station)
    station_names[iter_station] = iter_name

print(station_names)
```
